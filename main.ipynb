```python
# Import the pathlib and csv library
import numpy as np
from pathlib import Path
import csv

# Set the file path
budget_path = Path("../budget_data.csv")

# Initialize variable to hold budget
budget = []
numpy_nlist = []

# Initialize line_num variable
line_num = 0

# Open the input path as a file object
with open(budget_path, 'r') as csvfile:

    # Print the datatype of the file object
#    print(type(csvfile))

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    # Print the datatype of the csvreader
#    print(type(csvreader))

    # Go to the next row from the start of the file
    # (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    line_num += 1
    # Print the header
#    print(f"{header} <---- HEADER")

    # Read each row of data after the header
    for row in csvreader:
        # Print the row
#        print(row)
        # Set salary variable equal to the value in the 4th column of each row          str to int nested list
        budgets = [row[0],int(row[1])]
        # Append the row salary value to the list of salaries
        budget.append(budgets)
        
        # create nested list
        numpy_nlist.append(row)
        
# change string to int in nested list
#nested_list = [[head] + list(map(int, tail)) for head, *tail in numpy_nlist]

# Initialize metric variables
max_budgets = 0
min_budgets = 0
avg_budgets = 0
total_budgets = 0
count_budgets = 0

# Calculate the max, mean, and average of the list of salaries
for budgets in budget:

    # Sum the total and count variables  (can also use len(budgets))
    total_budgets += budgets[1]
    count_budgets += 1

    # Logic to determine min and max salaries
    if min_budgets == 0:
        min_budgets = budgets[1]
    elif budgets[1] > max_budgets:
        max_budgets = budgets[1]
    elif budgets[1] < min_budgets:
        min_budgets = budgets[1]

# Calculate the average salary, round to the nearest 2 decimal places
avg_budgets = round(total_budgets / count_budgets, 2)


# print list in nested list based on element in one of the lists
for sublist in budget:
    if sublist[1] != min_budgets:
        continue
    else:
        minimum_budget = sublist
        
# print list with max/min value in in it
for sublist in budget:
    if sublist[1] != max_budgets:
        continue
    else:
        maximum_budget = sublist

# Another way to print max or min list inside a nested list (remember it only works on int not str)
#max(numpy_df, key=lambda x: x[1])
#min(numpy_df, key=lambda x: x[1])

# Print the metrics
print("Financial Analysis")
print("----------------------------")
print(f'Total Months: {count_budgets}')
print(f'Total: ${total_budgets}')
print(f'Average Change: ${avg_budgets}')
print(f'Greatest Increase in Profits: {maximum_budget[0]} (${maximum_budget[1]})')
print(f'Greatest Decrease in Profits: {minimum_budget[0]} (${minimum_budget[1]})') 
```
