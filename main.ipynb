```python
"""
PyBank
I will be analysing the financial records of a company based on the dataset provided.
By Emiliano Amador
"""
# import all material needed
import numpy as np
from pathlib import Path
import csv

# Set the file path
csvpath = Path('budget_data.csv')

# Initialize variable to hold salaries
profits = []

# Initialize line_num variable
line_num = 0

# Open the input path as a file object
with open(csvpath, 'r') as csvfile:


    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')


    # Go to the next row from the start of the file
    # (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    line_num += 1
   
    
    # Read each row of data after the header
    for row in csvreader:
        
        
        profit = int(row[1])
        # Append the row 
        profits.append(profit)
```

```python
# Initialize metric variables
max_profit = 0
min_profit = 0
avg_profit = 0
total_profit = 0
count_profit = 0

# Calculate the max, mean, and average of the list of salaries
for profit in profits:

    # Sum the total and count variables
    total_profit += profit
    count_profit += 1

    # Logic to determine min and max salaries
    if min_profit == 0:
        min_profit = profit
    elif profit > max_profit:
        max_profit = profit
    elif profit < min_profit:
        min_profit = profit
      
        
# Calculate the average salary, round to the nearest 2 decimal places:
avg_profit = round(total_profit / count_profit, 2)

# Set the output header
header = ["Max_Profit", "Min_Profit", "Avg_Profit"]
# Create a list of metrics
metrics = [max_profit, min_profit, avg_profit]
```

```python
change_profit = np.diff(profits)
average_change_profit = np.mean(change_profit)
```

```python
max_change = 0
min_change = 0
total_change = 0

for change in change_profit:
    # Sum the total and count variables:
    total_change += change
    

    # Logic to determine min and max change:
    if min_change == 0:
        min_change = change
    elif change > max_change:
        max_change = change
    elif change < min_change:
        min_change = change
        
# format numbers:
average_change_formatted = round(average_change_profit, 2)
        
# Find the Date where the max_change and min_change occured:


# print it:
        
print("Financial Analysis")        
print("----------------------------")        
print (f"Total Months: {count_profit}")        
print (f"Total: ${total_profit}")        
print(f"Average Change: {average_change_formatted}")
print(f"Greatest Increase: ${max_change}")
print (f"Greatest Decrease: ${min_change}")




# Set the output file path:
output_path = Path("PyBank.txt")

# Open the output_path as a file object in "write" mode ('w')
# Write a header line and write the contents of 'text' to the file
with open(output_path, 'w') as file:
    file.write(
"Financial Analysis\n"
"----------------------------\n" )
    file.write(
        f"Total Months: {count_profit}\n"
        f"Total: ${total_profit}\n"
        f"Average Change: ${average_change_formatted}\n"
        f"Greatest Increase: ${max_change}\n"
        f"Greatest Decrease: ${min_change}\n"
    )
```
